#Class
Class is a blue-print to create an object.
Example:
car factory provide Body type of car, tyres, engine type etc.

Class contain 2 types of things-

1. Attributes - variable defined inside the class

2. Methods - function defined inside the class


```python
#class
class Animal:#class
  type = "cat" #attribute

  def sound(self):#method
    print("MewO!")

#Access class.attribute
print(Animal.type)

# To call an instance method, first create an instance of the class
a = Animal()
a.sound()
```

#Objects
Objects are instances of classes. Example- Bike, Ferrari

```python
class Fruits:#class
  name = "Apple"#attributes


  def details(self):#method
    print("I like fresh green apple")


#creating an object
A = Fruits()
#Accessing the attribute using objects
print(A.name)
#Accessing the method using object
A.details()
```

#Attribute & Methods
Attribute- there are 2 types of attributes

1. class attributes

2. Instance attributes

Methodes- there are 3 types of methods

1. Instance Method:

-Works with object (instance) data


-First parameter is always self


-Can access and modify instance attributes

2. Class Method:

-Uses @classmethod decorator

-First parameter is cls

-Can modify class variables

3. Static Method:

-Uses @staticmethod

-No self or cls

-Works like a normal function inside a class


```python
class Animal:
  name = "Lion" # class attribute

  def __init__(self, name):
    self.name = name #instance attribute

  def show(self):#Instance Method:
    print(f"Whats your favourite animal? {self.name}")

  @classmethod
  def hello(cls):#Class Method
    print(f"We are learning class methods")

  @staticmethod
  def stat(): #Static Method
    print("We are learning static methods")


obj = Animal("Tiger")
obj.show()#call instance method
obj.hello()#call class method
obj.stat()#call static method
```
