#List Slicing


```python
id = [101,105,103,108,106,67,89,35,78]
print(id[1:5])
print(id[:5])
print(id[5:])
print(id[5:-1])
print(id[:-3])
```

#List Method

```python
#Append- Add one element at the end of list
student2 = ['Casey', 25, 3.4,'cse']
print(student2)

#Append(value)- Add country as one element at the end of list
student2.append ("Canada")
print(student2)

#Insert(index, value)- replaceing values
student2.insert (-1, "EEE")
print(student2)
#Remove(value)- delete any value
student2. remove("EEE")
print(student2)

#Pop(index num) - delete any index
student2.pop(2)
print(student2)

#Reverse()- reverses in backword
student2.reverse()
print(student2)
```

```python

course_id= [23,2,56,90,56,89,23,1]
print(course_id)

#Sort() - Ascending order
course_id.sort()
print(course_id)

#descending order
course_id.sort(reverse=True)
print (course_id)

```

```python
num =[ 1,2,3,4,5]

#Copy()- make a shallow copy of list
num.copy()
print(num)

#reverse
num.reverse()
print(num)

#Clear()- delete all values
num.clear()
print(num)
```
